### Shuffling the order of the elements in a tensor


```python
not_shuffled = tf.constant([[1, 2],
                              [3, 4],
                              [6, 7]]) 
not_shuffled
```

```python
#shuffling the tensor 
tf.random.shuffle(not_shuffled)
# we can see that the tensor is shuffled about its first dimension'''

# we get random order each time we run it
```

```python
# what if we set seed value
tf.random.set_seed(42)     # This is the global level random seed
tf.random.shuffle(not_shuffled)

# we get the same sequence no matter how many times we run it
```

```python
# this way doesnt keep the sequence same
tf.random.shuffle(not_shuffled, seed = 42)   # This is the operation level random seed
```

If we want the shuffled tensor to be of same sequence and it doesnt change each time we run it then we must use the Global level tesnor  OR global along with the operation level

### Aggregating Tenors 

Aggregating means condensing tensors from multiple value to a smaller amount of value tensors 

```python
# Get the absoute value
E = tf.constant([-10, -7])
E
```

```python
# this gives the absolute value of a negative tensor
tf.abs(E)
```

Lets go through the following form of aggregation 
* Get the making
* Get the minimum
* Get the mean of the tensor
* Get the sum of the tensor


```python
# Creating a random tensor
F = tf.constant(np.random.randint(0, 100, size = 50))
F
```

```python
# Finding the minimum of the 
'''it is similar to numpy where we use np.min(F) but in tensorflow there is a convention to use reduce_min'''

tf.reduce_min(F)
```

```python
# Finding the Max
tf.reduce_max(F)
```

```python
# Finding the Sum
tf.reduce_sum(F)
```

```python
# Finding the mean
tf.reduce_mean(F)
```

```python
# Finding the Variance
tf.reduce_std(F) # this wont work

'''tf.reduce_variance''' # this will also not work
```

```python
# finding the variance
import tensorflow_probability as tfp
tfp.stats.variance(F)
```

```python
#finding the standard deviation
F = tf.cast(F, dtype = tf.float16)
tfp.stats.stddev(F)
```

```python
F = tf.cast(F, dtype = tf.float32)
tf.math.reduce_std(F)
```

```python
# better approach instead of writing F seperately
tf.math.reduce_std(tf.cast(F, dtype = tf.float32)) 
```

```python
tf.math.reduce_variance(F)
```

### Finding the positional minimun and maximum 

```python
G = tf.random.set_seed(40)
G = tf.random.uniform(shape = [50])
G
```

```python
# to find the positional maximum
tf.argmax(G) # position where there is max value element
```

```python
# finding the value on the index we got max value
G[tf.argmax(G)]
```

```python
#the max element in the tensor
tf.reduce_max(G)
```

```python
# checking weather the two are equal or not
G[tf.argmax(G)] == tf.reduce_max(G)
```

```python
# to find the positional minimum
tf.argmin(G)
```

```python
# finding the value on the index we got min value
G[tf.argmin(G)]
```

### One hot encodin of tensors
when we have words as input for a feature then we use one hot encoding like it turn features like red blue green too 1 and 0 

```python
some_list = [0,1,2,3]
tf.one_hot(some_list, depth = 4) # depth as for a 4x4 we have 4 elements and we want to get a 4x4 matrix
```

```python
'''
      [[1., 0., 0., 0.],  <-- this row depicts the number 0
       [0., 1., 0., 0.],  <-- this row depicts the number 1
       [0., 0., 1., 0.],  <--this row depicts the number 2
       [0., 0., 0., 1.]]  <--this row depicts the number 3
       '''
```

```python
tf.one_hot(some_list, depth = 4 , on_value = 'hello', off_value= 'why') # if we wish to use any other value instead of 1 or 0 we thus there use on_value and off_value
```
